```python
import os
import glob
import pandas as pd
```

```python
def read_config_file(file_path):
    config = {}
    with open(file_path, 'r') as file:
        for line in file:
            key, value = line.strip().split(': ')
            config[key] = value
    return config
```

```python
patients = os.listdir('ACDC_raw/database/testing')
#patients.remove('.ipynb_checkpoints')
patients.sort()
```

```python
patient_info = pd.DataFrame(index = patients, columns = ['ED_frame', 'ES_frame', 'Group', 'Height', 'Weight', 'NbFrame'])
```

```python
patient_info.index.name = 'patient'
```

```python
for patient in patients:
    file_path = os.path.join( 'ACDC_raw/database/testing', patient, 'Info.cfg')
    try:
        config = read_config_file(file_path)
    except NotADirectoryError:
        continue
    patient_info['ED_frame'][patient] = config['ED']
    patient_info['ES_frame'][patient] = config['ES']
    patient_info['Group'][patient] = config['Group']
    patient_info['Height'][patient] = config['Height']
    patient_info['Weight'][patient] = config['Weight']
    patient_info['NbFrame'][patient] = config['NbFrame']
    
```
